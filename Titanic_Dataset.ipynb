```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('/content/Titanic Dataset.csv')
```

```python
# Basic Info
print("Basic Info:")
print(df.info())
print("\nFirst 5 Rows:")
print(df.head())
```

```python
# Summary Statistics
print("\nSummary Statistics:")
print(df.describe(include='all'))
```

```python
# Check for Missing Values
plt.figure(figsize=(10, 4))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis", yticklabels=False)
plt.title("Missing Data Heatmap")
plt.show()
```

```python
# Count of Missing Values
print("\nMissing Values per Column:")
print(df.isnull().sum())
```

```python
# EDA Visualizations
plt.figure(figsize=(18, 12))
```

```python
# 1. Survival Count
plt.subplot(2, 3, 1)
sns.countplot(data=df, x='Survived', palette='Set2')
plt.title('Survival Count (0 = No, 1 = Yes)')
```

```python
# 2. Survival by Sex
plt.subplot(2, 3, 2)
sns.countplot(data=df, x='Sex', hue='Survived', palette='Set1')
plt.title('Survival by Sex')
```

```python
# 3. Survival by Pclass
plt.subplot(2, 3, 3)
sns.countplot(data=df, x='Pclass', hue='Survived', palette='Set3')
plt.title('Survival by Pclass')
```

```python
# 4. Age Distribution
plt.subplot(2, 3, 4)
sns.histplot(data=df, x='Age', kde=True, bins=30, color='skyblue')
plt.title('Age Distribution')
```

```python
# 5. Fare Distribution
plt.subplot(2, 3, 5)
sns.histplot(data=df, x='Fare', kde=True, bins=40, color='salmon')
plt.title('Fare Distribution')
```

```python
# 6. Correlation Heatmap
plt.subplot(2, 3, 6)
corr = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
```

```python
plt.tight_layout()
plt.show()
```
